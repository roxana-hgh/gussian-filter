## Gaussian Filter

In image processing, a Gaussian Blur is utilized to reduce the amount of noise in an image

the Gaussian blurring algorithm will scan over each pixel of the image, and recalculate the pixel value based on the pixel values that surround it.

The area that is scanned around each pixel is called the kernel. A larger kernel scans a larger amount of pixels that surround the center pixel.

Gaussian blurring doesnâ€™t weigh each pixel equally, however. The closer a pixel is to the center, the greater it affects the weighted average used to calculate the new center pixel value. The image below demonstrates this function. This method assumes pixels closest to the center pixel will be closest to the true value of that pixel, so they will influence the averaged value of the center pixel greater than pixels further away.

```python
from IPython.display import Image
display(Image(filename='Screenshot_1.png'))
```

The size of the sigma of the function dictates how wide the curve should be inside the kernel, in the same way
 that sigma, or standard deviation, of a Normal Distribution curve will dictate the width of the curve. 

The openCV GaussianBlur() function takes in 3 parameters here: the original image, the kernel size, and the sigma.

```python
import cv2
from matplotlib import pyplot as plt

# Read the imaage
img = cv2.imread('noise.jpg')

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Apply Gaussian blur with kernel size (x,y) and sigma value of x (if sigma = 0 -> sigma = 0.3*((ksize-1)*0.5 - 1) + 0.8)
gaussian = cv2.GaussianBlur(img, (3,3), 1)
gaussian2 = cv2.GaussianBlur(img, (3,3), 0.2)
gaussian3 = cv2.GaussianBlur(img, (3,3), 50)
gaussian4 = cv2.GaussianBlur(img, (9,9), 1)
gaussian5 = cv2.GaussianBlur(img, (9,9), 0.2)
gaussian6 = cv2.GaussianBlur(img, (9,9), 50)
gaussian7 = cv2.GaussianBlur(img, (21,21), 1)
gaussian8 = cv2.GaussianBlur(img, (21,21), 0.2)
gaussian9 = cv2.GaussianBlur(img, (21,21), 50)

fig, axs = plt.subplots(3, 3,figsize=(15, 15))
axs[0,0].imshow(cv2.cvtColor(gaussian, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('kernel: 3 x 3, Sigma: 1')

axs[0,1].imshow(cv2.cvtColor(gaussian2, cv2.COLOR_BGR2RGB))
axs[0, 1].set_title('kernel: 3 x 3, Sigma: 0.2')

axs[0,2].imshow(cv2.cvtColor(gaussian3, cv2.COLOR_BGR2RGB))
axs[0, 2].set_title('kernel: 3 x 3, Sigma: 50')

axs[1,0].imshow(cv2.cvtColor(gaussian4, cv2.COLOR_BGR2RGB))
axs[1, 0].set_title('kernel: 9 x 9, Sigma: 1')

axs[1,1].imshow(cv2.cvtColor(gaussian5, cv2.COLOR_BGR2RGB))
axs[1, 1].set_title('kernel: 9 x 9, Sigma: 0.2')

axs[1,2].imshow(cv2.cvtColor(gaussian6, cv2.COLOR_BGR2RGB))
axs[1, 2].set_title('kernel: 9 x 9, Sigma: 50')

axs[2,0].imshow(cv2.cvtColor(gaussian7, cv2.COLOR_BGR2RGB))
axs[2, 0].set_title('kernel: 21 x 21, Sigma: 1')

axs[2,1].imshow(cv2.cvtColor(gaussian8, cv2.COLOR_BGR2RGB))
axs[2, 1].set_title('kernel: 21 x 21, Sigma: 0.2')

axs[2,2].imshow(cv2.cvtColor(gaussian9, cv2.COLOR_BGR2RGB))
axs[2, 2].set_title('kernel: 21 x 21, Sigma: 50')

```
